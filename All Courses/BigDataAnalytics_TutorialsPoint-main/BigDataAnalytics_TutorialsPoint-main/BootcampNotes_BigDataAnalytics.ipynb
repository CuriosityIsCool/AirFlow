## Big Data Analytics - Data Life Cycle


### Traditional Data Mining Life Cycle

#### CRISP-DM Methodology

The CRISP-DM methodology that stands for Cross Industry Standard Process for Data Mining, is a cycle that describes commonly used approaches that data mining experts use to tackle problems in traditional BI data mining. It is still being used in traditional BI data mining teams.

image.png

Big data analytics is the process of analyzing large and complex data sets to uncover hidden patterns, correlations, and insights. The CRISP-DM methodology is a widely-used framework for data mining projects. 

It consists of six stages: business understanding, data understanding, data preparation, modeling, evaluation, and deployment. 

* In the business understanding stage, project objectives are defined and a preliminary plan is developed. 
* In the data understanding stage, data is collected and examined for quality issues and insights. 
* In the data preparation stage, data is cleaned and transformed for modeling. 
* In the modeling stage, various modeling techniques are applied and calibrated for optimal values. 
* In the evaluation stage, the model is thoroughly evaluated to ensure it achieves the business objectives. 
* Finally, in the deployment stage, the model is presented in a useful way to the customer, who will carry out the necessary actions to make use of it.

#### SEMMA Methodology
SEMMA is a data mining modeling methodology developed by SAS, and it stands for Sample, Explore, Modify, Model, and Assess. SEMMA focuses on the modeling aspect of data mining. 

* The process begins with data sampling and partitioning in the Sample phase. 
* The Explore phase involves visualizing data to understand relationships and abnormalities. 
* In the Modify phase, variables are selected, created, and transformed in preparation for modeling. 
* The Model phase applies various modeling techniques to create models that provide the desired outcome. 
* The final stage, Assess, evaluates the modeling results for reliability and usefulness.

The main difference between CRISP-DM and SEMMA is that CRISP-DM places more importance on stages prior to modeling, such as understanding the business problem and preprocessing the data. In contrast, SEMMA focuses more on the modeling stage itself.

### Big Data Life Cycle

The life cycle consists of various stages, including `Business Problem Definition`, `Research`, `Human Resources Assessment`, `Data Acquisition`, `Data Munging`, `Data Storage`, `Exploratory Data Analysis`, `Data Preparation for Modeling and Assessment`, `Modeling`, and `Implementation`. 

The `Business Problem Definition` stage involves defining the problem and assessing the potential gains and costs of the project. In the `Research` stage, solutions that are reasonable for the company are analyzed by looking at what other companies have done in the same situation, and a methodology for the future stages is defined. In the `Human Resources Assessment` stage, the current staff's ability to complete the project successfully is analyzed to determine whether there is a need to outsource a part of the project or hire more people. 

`Data Acquisition` is a critical stage in the big data life cycle that defines which type of profiles would be needed to deliver the resultant data product. Data gathering involves gathering unstructured data from different sources, which can be a non-trivial step of the process. The `Data Munging` stage involves storing the retrieved data in an easy-to-use format. 

In the `Data Storage` stage, the data is sometimes stored in a database, and big data technologies offer a variety of alternatives for this stage. In the `Exploratory Data Analysis` stage, the data exploration phase is mandatory to understand the data, which is done with statistical techniques and plotting the data. The Data Preparation for `Modeling and Assessment` stage involves reshaping the cleaned data retrieved previously and using statistical preprocessing for missing values imputation, outlier detection, normalization, feature extraction, and feature selection.

The `Modeling` stage involves trying different models and selecting the best model or combination of models, evaluating their performance on a left-out dataset. Finally, the `Implementation` stage involves implementing the data product developed in the data pipeline of the company and setting up a validation scheme to track its performance.

## Big Data Analytics - Core Deliverables

Data products that result from developing a big data product are in most of the cases some of the following 

* Machine learning implementation  This could be a classification algorithm, a regression model or a segmentation model.

* Recommender system  The objective is to develop a system that recommends choices based on user behavior. `Netflix` is the characteristic example of this data product, where based on the ratings of users, other movies are recommended.

* Dashboard  Business normally needs tools to visualize aggregated data. A dashboard is a graphical mechanism to make this data accessible.

* Ad-Hoc analysis  Normally business areas have questions, hypotheses or myths that can be answered doing ad-hoc analysis with data.

## Big Data Analytics - Data Scientist


The role of a data scientist is normally associated with tasks such as predictive modeling, developing segmentation algorithms, recommender systems, A/B testing frameworks and often working with raw unstructured data.

Here is a set of skills a data scientist normally need to have 

* Programming in a statistical package such as: R, Python, SAS, SPSS, or Julia
* Able to clean, extract, and explore data from different sources
* Research, design, and implementation of statistical models
* Deep statistical, mathematical, and computer science knowledge

```python
import ipywidgets as widgets

options = ['Programming in a statistical package (e.g. R, Python, SAS, SPSS, Julia)', 
           'Able to clean, extract, and explore data from different sources', 
           'Research, design, and implementation of statistical models', 
           'Deep statistical, mathematical, and computer science knowledge']
checkboxes = []

for option in options:
    checkbox = widgets.Checkbox(value=False, description=option)
    checkboxes.append(checkbox)
    display(checkbox)
```

A data architect defines the tools and the architecture the data would be stored at, whereas a data scientist uses this architecture. Of course, a data scientist should be able to set up new tools if needed for ad-hoc projects, but the infrastructure definition and design should not be a part of his task.
