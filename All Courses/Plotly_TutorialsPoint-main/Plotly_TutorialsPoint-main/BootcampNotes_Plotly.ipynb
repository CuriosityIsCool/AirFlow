## Bar Charts (Part 1)

````python
# getting data
import requests
animals_url = "https://gist.githubusercontent.com/borlaym/585e2e09dd6abd9b0d0a/raw/6e46db8f5c27cb18fd1dfa50c7c921a0fbacbad0/animals.json"
animals_list = requests.get(animals_url).json()

num_of_animals = tuple(random.randint(10,60) for x in range(len(animals_list)))  # random number of animals

print(animals_list)
print(num_of_animals)
````

````python
# Plotting now
import plotly.graph_objects as go
fig = go.Figure(go.Bar(x=animals_list,
                       y = num_of_animals))   # created graph obj
fig.show()
````

````python
# adding trace
import plotly.graph_objects as go
fig = go.Figure([go.Bar(x=animals_list, y = num_of_animals, name="Plot1"),
                 go.Bar(x=animals_list, y = sorted(num_of_animals), name="Plot1sorted")])   # created graph obj
fig.show()
````

above is `barmode = "group"` we can use stack barmode to stack graphs in `fig.update_layout(barmode=)` above each other

## Pie Charts
just use `go.Pie`

````python
# creating data to plot
percentages=(78.08, 20.95, 0.93)
gases = ("nitrogen", "oxygen", "argon")
````

````python
import plotly.graph_objects as go
fig = go.Figure(go.Pie(labels = gases, 
                       values = percentages,
                       textinfo = "value+label"))
fig.show()
````

### pulling out a piece

````python
import plotly.graph_objects as go
fig = go.Figure(go.Pie(labels = gases, 
                       values = percentages,
                       textinfo = "value+label",
                       pull=[0,0.1,0.2,0.5]))
fig.show()
````

### Adding hole (donut chart)

````python
import plotly.graph_objects as go
fig = go.Figure(go.Pie(labels = gases, 
                       values = percentages,
                       textinfo = "value+label",
                       pull=[0,0.1,0.2,0.5],
                       hole = 0.2)) # 0.2 means 20% hole
fig.show()
````

## Plotly - 3D Scatter and Surface Plot

````python
import plotly.graph_objs as go
import numpy as np
from plotly.offline import iplot

z = np.linspace(0, 10, 50)
x = np.cos(z)
y = np.sin(z)
trace = go.Scatter3d(
   x = x, y = y, z = z,mode = 'markers', marker = dict(
      size = 12,
      color = z, # set color to an array/list of desired values
      colorscale = 'Viridis'
      )
   )
layout = go.Layout(title = '3D Scatter plot')
fig = go.Figure(data = [trace], layout = layout)
iplot(fig)
````
