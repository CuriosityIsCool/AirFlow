* ALTER
  ALTER in PostgreSQL is a command used to modify the structure of existing database objects, such as tables or columns.

  Here's a simple example:

  Let's say you have a table called "employees" with the following columns: "id", "name", "age", and "salary". If you want to add a new column called "department" to the table, you can use the ALTER command as follows:

  ```
  ALTER TABLE employees ADD COLUMN department TEXT;
  ```

  This command adds a new column called "department" with a data type of TEXT to the "employees" table. Now you can store the department information for each employee in this new column.

  This link basically describes a set of commands in SQL (Structured Query Language) that can be used to modify or change the structure and properties of different database objects such as tables, functions, triggers, indexes, etc.

For example, the "ALTER TABLE" command can be used to add or remove columns from a table, rename a table, or change the owner of a table. The "ALTER FUNCTION" command can be used to modify the definition of a function, such as its input parameters or return type. The "ALTER USER" command can be used to change the properties of a user account in the database.

Overall, these commands can be useful for managing and updating databases to ensure that they meet the needs of different applications and users.

## PostgreSQL - LIKE Clause

WHERE part having different LIKE clause with '%' and '_' operators 

* `_` represents filled position (jitne `_` utni position me koi char)
* `%` represents any combination of char (% matlab `ab` ya `abc` ya `1537` kuch bhi size ke char ho skate hain)

`WITH` created temporary variables

Example:
```
WITH moved_rows AS (
   DELETE FROM COMPANY
   WHERE
      SALARY >= 30000
   RETURNING *
)
INSERT INTO COMPANY1 (SELECT * FROM moved_rows);
```
* `moved_rows` is temporary variable created in above example

## PostgreSQL - HAVING Clause
The HAVING clause allows us to pick out particular rows where the function's result meets some condition.

The WHERE clause places conditions on the selected columns, whereas the HAVING clause places conditions on groups created by the GROUP BY clause.


Jaise `SELECT` ke sath `WHERE` aata hai, waise hi `GROUPBY` ke sath `HAVING` aata hai.....(conditions checking ke liye)

## Case Expression in PostgreSQL

Example Code 1:
```
SELECT title,
       length,
       CASE
           WHEN length> 0 AND length <= 50 
                THEN 'Short'
           WHEN length > 50 AND length <= 120 
                THEN 'Medium'
           WHEN length> 120 
                THEN 'Long'
       END duration
FROM film
ORDER BY title;
```

image.png


Example code 2: 
```
SELECT
	SUM (CASE
          WHEN rental_rate = 0.99 
            THEN 1
	        ELSE 0
	    END
	    ) AS "Economy",
	SUM (
		CASE
		  WHEN rental_rate = 2.99 
        THEN 1
		  ELSE 0
		END
	  ) AS "Mass",
	SUM (
		CASE
		  WHEN rental_rate = 4.99 
        THEN 1
		  ELSE 0
		END
	    ) AS "Premium"
FROM
	film;
```

## Functions in PostgreSQL

```
create or replace function count_emails() returns integer as $total_emails$
  declare total_emails integer;
  begin
    select count(emailColumn) into total_emails from employeesTable;
    return total_emails;
  end;

$total_emails$ language plpgsql;
```

Running Function:
```
select count_emails();
```
