## Statistics - Analysis of Variance (ANOVA)

* One-way ANOVA have only one independent variable and refers to numbers in this variable

* Two way ANOVA uses two independent variables. Two-way ANOVAs are also termed as factorial ANOVA and can be balanced as well as unbalanced. Balanced refers to having same number of participants in each group where as unbalanced refers to having different number of participants in each group.
  * Hierarchical approach(Type 1) -If data was not intentionaly unbalanced and has some type of hierarchy between the factors.

  * Classical experimental approach(Type 2) - If data was not intentionaly unbalanced and has no hierarchy between the factors.

  * Full Regression approach(Type 3) - If data was intentionaly unbalanced because of population.

* N-way or Multivariate ANOVA - N-way ANOVA have multiple independent variables.

### ANOVA Test Procedure
Following are the general steps to carry out ANOVA.

1. Setup null and alternative hypothesis where null hypothesis states that there is no significant difference among the groups. And alternative hypothesis assumes that there is a significant difference among the groups.

2. Calculate F-ratio and probability of F.

3. Compare p-value of the F-ratio with the established alpha or significance level.

4. If p-value of F is less than 0.5 then reject the null hypothesis.

5. If null hypothesis is rejected, conclude that mean of groups are not equal.

## Formulas

* Chi-squared Distribution: 

The probability density function (pdf) of the chi-squared distribution with k degrees of freedom is:

`f(x) = 1/2^(k/2) * x^(k/2 - 1) * e^(-x/2) / Gamma(k/2)`

where `Gamma` is the gamma function.

* Chi Squared table:

A Chi Squared table shows the critical values for the chi-squared distribution for various degrees of freedom and levels of significance.

* Circular Permutation:

The formula for the number of circular permutations of `n distinct objects` is `(n-1)!`.

* Cluster sampling:

The formula for the sample mean for a cluster sample is the sum of the means of each cluster divided by the number of clusters.

* Cohen's kappa coefficient:

The formula for Cohen's kappa coefficient is:

`kappa = (P(A) - P(E)) / (1 - P(E))`

where `P(A)` is the observed agreement and `P(E)` is the expected agreement.

* Combination:

The formula for the number of combinations of `n` objects taken `r` at a time is:

`C(n, r) = n!/r!(n-r)!`

* Combination with replacement:

The formula for the number of combinations of `n` objects taken `r` at a time with replacement is:

`C(n+r-1, r) = (n+r-1)!/(r!(n-1)!)`

* Continuous Uniform Distribution:

The probability density function of the continuous uniform distribution on the interval `[a,b]` is:

`f(x) = 1/(b-a) for a ≤ x ≤ b and 0 otherwise`

* Continuous Series Arithmetic Mean:

The formula for the arithmetic mean of a continuous series is:

`mean = (∑(f * X))/n`

where `f` is the frequency of each class, `X` is the midpoint of each class, and `n` is the total number of observations.

* Continuous Series Arithmetic Median:

The formula for the arithmetic median of a continuous series is:

`median = L + ((n/2 - F)/f) * w`

where `L` is the lower class boundary of the median class, `F` is the cumulative frequency up to the median class, `f` is the frequency of the median class, `n` is the total number of observations, and `w` is the width of each class.

* Continuous Series Arithmetic Mode:

The formula for the arithmetic mode of a continuous series is:

`mode = L + ((f1-f0)/(2f1-f0-f2)) * w`

where `L` is the lower class boundary of the modal class, `f1` is the frequency of the modal class, `f0` is the frequency of the class preceding the modal class, `f2` is the frequency of the class succeeding the modal class, and `w` is the width of each class.

* Cumulative Frequency:

The cumulative frequency of a class is the sum of the frequencies of that class and all previous classes.

* Co-efficient of Variation:

The formula for the coefficient of variation is:

`CV = (standard deviation / mean) * 100%`

* Correlation Co-efficient:

The formula for the correlation coefficient between two variables `X` and `Y` with n observations is:

`r = (∑XY - (1/n) * (∑X) * (∑Y)) / (sqrt(∑X^2 - (1/n) * (∑X)^2) * sqrt(∑Y^2 - (1/n) * (∑Y)^2))`

* Probability Bayes Theorem:

P(A|B) = P(B|A) * P(A) / P(B)

* Probability Density Function:

Probability density function, f(x), of a continuous random variable X, is a function that describes the relative likelihood for this random variable to take on a given value x.

* Process Capability (Cp) & Process Performance (Pp):

Cp = (USL - LSL) / (6*sigma), where:
USL - Upper Specification Limit
LSL - Lower Specification Limit
sigma - standard deviation of the process

Pp = (USL - LSL) / (6*sqrt(MSE)), where:
MSE - mean squared error of the process

* Process Sigma:

Process sigma represents the standard deviation of a process. A higher value of process sigma means less variability and greater process stability.

* Qualitative Data Vs Quantitative Data:

Qualitative data is non-numerical data that is descriptive in nature, such as colors, shapes, and names. Quantitative data is numerical data that is measurable, such as heights, weights, and ages.

* Quartile Deviation:

Quartile deviation is defined as half the difference between the first and third quartiles, i.e. QD = (Q3 - Q1)/2.

* Range Rule of Thumb:

Range rule of thumb states that the range of a data set can be estimated as four times the standard deviation.
