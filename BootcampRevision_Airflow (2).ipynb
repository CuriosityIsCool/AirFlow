## Creating Airflow Scheduler (runs parallel with airflow server)
In order to run the `DAG`s we have to start the Airflow Scheduler

## Problem
I have to run `airflow scheduler` in parallel to `airflow server`

## Solution can be
```
import multiprocessing

def task1():
  # your code for task1

def task2():
  # your code for task2

if __name__ == '__main__':
  p1 = multiprocessing.Process(target=task1)
  p2 = multiprocessing.Process(target=task2)
  
  p1.start()
  p2.start()
  
  p1.join()
  p2.join()

```

## Solution is just adding `&`

````python
print(ngrok.connect(port_no).public_url)
````

````python
!airflow scheduler & airflow webserver -p 5002  # this needs to be kept running at same port as ngrok
````

## Creating First DAG

`DAG` is stored in `/dags` folder as a python file

````python
!ls
````

````python
!ls
!mkdir dags
!touch dags/first_Dag_try.py
````

````python
python_code = '''
from datetime import datetime, timedelta

from airflow import DAG
from airflow.operators.bash import BashOperator


default_args = {
    'owner': 'aditya',
    'retries': 5,
    'retry_delay': timedelta(minutes=2)
}


with DAG(
    dag_id='our_first_dag_v5',
    default_args=default_args,
    description='This is our first dag that we write',
    start_date=datetime(2021, 7, 29, 2),
    schedule_interval='@daily'
) as dag:
    task1 = BashOperator(
        task_id='first_task',
        bash_command="echo hello world, this is the first task!"
    )

    task2 = BashOperator(
        task_id='second_task',
        bash_command="echo hey, I am task2 and will be running after task1!"
    )

    task3 = BashOperator(
        task_id='thrid_task',
        bash_command="echo hey, I am task3 and will be running after task1 at the same time as task2!"
    )

    # Task dependency method 1
    # task1.set_downstream(task2)
    # task1.set_downstream(task3)

    # Task dependency method 2
    # task1 >> task2
    # task1 >> task3

    # Task dependency method 3
    task1 >> [task2, task3]
'''

with open("/content/dags/first_Dag_try.py", "w") as f:
  f.write(python_code)
````
